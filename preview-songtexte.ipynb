```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Basic info about the dataset
print(df.info())

# Descriptive statistics
print(df.describe())

# Count of unique values in 'tag' column which seems to be the genre
print(df['tag'].value_counts())

# Visualize the distribution of views
plt.figure(figsize=(10, 5))
sns.histplot(df['views'], bins=50)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Number of Songs')
plt.show()

# Visualize the number of songs per year
plt.figure(figsize=(15, 5))
sns.countplot(x='year', data=df)
plt.title('Number of Songs per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=90)
plt.show()


# Boxplot to see the distribution of views by genre
plt.figure(figsize=(15, 5))
sns.boxplot(x='tag', y='views', data=df)
plt.title('Views Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Views')
plt.xticks(rotation=45)
plt.show()
```
