```python
import pandas as pd
df = pd.read_csv("/kaggle/input/genius-song-lyrics-with-language-information/song_lyrics.csv")
df.head()
```

### Filter Data 

```python
import pandas as pd

year_df=df[df["year"]>=1960]
english_df = year_df[year_df['language'] == 'en']
english_df=english_df[english_df['tag'] != 'misc']

# Now, sample 20,000 random rows from the English songs
random_english_df = english_df.sample(n=50000)

# Export the random rows to a new CSV file
random_english_df.to_csv("sampled_english_songs_100000.csv", index=False)


```

```python
# Re-create the original filtering steps for English songs since 1960 without the 'misc' tag
year_df = df[df["year"] >= 1960]
english_df = year_df[year_df['language'] == 'en']
english_df = english_df[english_df['tag'] != 'misc']

# Now, sample 20,000 random rows from the English songs
df_1 = english_df.sample(n=50000)

# Exclude the rows that are already in the random_english_df
remaining_df = english_df[~english_df.index.isin(df_1.index)]

# Sample 10,000 random rows from the remaining DataFrame
final_test_df = remaining_df.sample(n=10000)
final_test_df.head()
```

```python
# Find the minimum size among all categories
min_size = english_df['tag'].value_counts().min()

# Sample from each category
sampled_dfs = [df_group.sample(n=10000) for tag, df_group in english_df.groupby('tag')]

# Combine the samples into one DataFrame
equal_sample_df = pd.concat(sampled_dfs)
```

```python
df_1.to_csv("sampled_english_songs_50000.csv", index=False)
final_test_df.to_csv("final_test_english_songs_10000.csv", index=False)
equal_sample_df.to_csv("equal_sample_english_songs_50000.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Basic info about the dataset
print(df.info())

# Descriptive statistics
print(df.describe())

# Count of unique values in 'tag' column which seems to be the genre
print(df['tag'].value_counts())

# Visualize the distribution of views
plt.figure(figsize=(10, 5))
sns.histplot(df['views'], bins=50)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Number of Songs')
plt.show()

# Visualize the number of songs per year
plt.figure(figsize=(15, 5))
sns.countplot(x='year', data=df)
plt.title('Number of Songs per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=90)
plt.show()


# Boxplot to see the distribution of views by genre
plt.figure(figsize=(15, 5))
sns.boxplot(x='tag', y='views', data=df)
plt.title('Views Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Views')
plt.xticks(rotation=45)
plt.show()
```
